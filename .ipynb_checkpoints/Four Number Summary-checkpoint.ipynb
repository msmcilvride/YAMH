```python
import numpy as np
import random as rd
```

```python
def d3():
    return rd.randint(1, 3)

def d6():
    return rd.randint(1, 6)
```

```python
def to_hit(BS, Expert_Crafters=False):
    
    results = []

    for _ in range(10000):

        die_roll = d6()

#         print("Hit roll:", hit_roll)

        if die_roll < BS:
            results.append(0)
#             print("Data: 0")
            continue
        else:
            results.append(1)
#             print("Data: 1")
            pass
    
    print("Hits")
    print("   Sum:", sum(results))
    print("  Mean:", np.mean(results))
    
    return np.mean(results)

def to_wound(threshold):
    
    results = []

    for _ in range(10000):

        die_roll = d6()

#         print("Hit roll:", hit_roll)

        if die_roll < threshold:
            results.append(0)
#             print("Data: 0")
            continue
        else:
            results.append(1)
#             print("Data: 1")
            pass

    print("Wounds")
    print("   Sum:", sum(results))
    print("  Mean:", np.mean(results))

    return np.mean(results)


def save_roll(Sv, AP, invuln=99):
    
    results = []
    
    # If the AP would reduce the armor to less
    # than the invuln, then use the invuln save
    # and ignore AP.
    if Sv - AP > invuln:
        Sv = invuln
        AP = 0

    for _ in range(10000):

        die_roll = d6() + AP

#         print("Die roll:", die_roll)

        if die_roll >= Sv:
            results.append(0)
#             print("Result: 0")
            continue
        else:
            results.append(1)
#             print("Result: 1")
            pass
    
    print("Failed Saves")
    print("   Sum:", sum(results))
    print("  Mean:", np.mean(results))
    
    return np.mean(results)

def damage_roll():
    
    results = []
    
    for _ in range(10000):
        
        damage_roll = d6()
        
        results.append(damage_roll)
        
    print("Damage")
    print("   Sum:", sum(results))
    print("  Mean:", np.mean(results))
        
    return np.mean(results)
```

```python
hits = to_hit(3)
wounds = to_wound(3)
failed_saves = save_roll(3, -4, invuln=99)
damage = damage_roll()
print("")
print("Average Damage:", hits * wounds * failed_saves * damage)
```
